```python
import pandas as pd
import os 
import glob
```

```python
path = 'Data' 
all_files = sorted(glob.glob(os.path.join(path, '*.txt')))
data_list = []

for eachfile in all_files: 
    data = open(eachfile, "r")
    for line in data.readlines(): #each list item is each line in file
        linebreak = line.split('\t') #removes \
        data_list.append(linebreak) #append next file as list item 
        
second_list = data_list [:2] #takes first two items of data_list (header + data row)
for num in range (3, 98, 2): #start from 4th item (data of file 2), go up by increments of 2(skip header of every file)
    second_list.append(data_list[num])

all_df = pd.DataFrame(second_list[1:], columns = second_list[0])
print(all_df)
```

```python
sequence_df = all_df.iloc[::, 0:3]
sequence_df
```

```python
'''
finding whether the gene has point mutation or not
'''
x = 0 
name_list = []
mutation_list = []
difference_list = []
for x in range (len(sequence_df)): 
    WT = sequence_df.iloc[x, 1]
    M = sequence_df.iloc[x, 2]
    name = sequence_df.iloc[x,0]
    if len(WT) < len(M): 
        difference1 = len(M)-len(WT)
        name_list.append(name)
        mutation_list.append('Insertion')
        difference_list.append(difference1)
    elif len(WT) > len(M): 
        difference2 = len(WT)-len(M)
        name_list.append(name)
        mutation_list.append('Deletion')
        difference_list.append(difference2)
    elif len(WT) == len(M): 
        name_list.append(name)
        mutation_list.append('Substitution')
        difference_list.append(0)
    x += 1

mutation_dict = {
    'Gene': name_list, 'Mutation': mutation_list, 'Hamming': difference_list}
#print (name_list)
#print (mutation_list)
#print (difference_list)
mutation_df = pd.DataFrame.from_dict(mutation_dict)
print(mutation_df)
```

```python
mutation_counts = mutation_df.groupby('Mutation').size()
#group by column 'Mutation' and count occurences
print (mutation_counts)
```

COMBINING everything together to find mean difference 

```python
WTmRNA = meanmRNA_WT['Mean.WT.mRNA']
MmRNA = meanmRNA_M['Mean.Mut.mRNA']
WTp = meanprotein_WT['Mean.WT.protein']
Mp = meanprotein_M['Mean.M.protein']
WTcv = meanCV_WT['Mean.WT.Viability']
Mcv = meanCV_M['Mean.M.Viability']
mut_type = mutation_df[['Gene','Mutation']]

frames = [mut_type, WTmRNA, MmRNA, WTp, Mp, WTcv, Mcv]
mean_values = pd.concat(frames, axis=1) #axis = 1 = columns
print (mean_values)
```

Add 3 columns detailing the difference between mean mRNA/protein/cell viability for mutant vs WT

```python
mean_values['mRNA.Difference'] = mean_values['Mean.WT.mRNA'].sub(mean_values['Mean.Mut.mRNA'])
mean_values['protein.Difference'] = mean_values['Mean.WT.protein'].sub(mean_values['Mean.M.protein'])
mean_values['CellViability.Difference'] = mean_values['Mean.WT.Viability'].sub(mean_values['Mean.M.Viability'])
print (mean_values)
```

Seeing whether the mutation has caused the mRNA expression, protein expression, and cell viability to increase, decrease, or no difference

```python
lessormore_mRNA = mean_values['Mean.WT.mRNA'] > mean_values['Mean.Mut.mRNA']
lessormore_protein = mean_values['Mean.WT.protein'] > mean_values['Mean.M.protein']
lessormore_CV = mean_values['Mean.WT.Viability'] > mean_values['Mean.M.Viability']
mut_type = mutation_df[['Gene','Mutation']]
lessormore_frame = [mut_type, lessormore_mRNA, lessormore_protein, lessormore_CV]


lessormore_data = pd.concat(lessormore_frame, axis=1)
lessormore_data.columns = ['Gene', 'Mutation','WT > M mRNA', 'WT > M protein', 'WT > M CV']
print(lessormore_data)
```
